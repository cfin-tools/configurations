```python
%%dot -f svg

digraph G {
    rankdir=TB;
    raw -> maxfilter -> "visualise &\ncompare" -> problems;
    raw -> "visualise &\ncompare";
    problems -> yes -> "refine parameters" -> maxfilter;
    problems -> no -> "save raw_(t)sss";    
}
```

```python
%%dot -f svg

digraph G {
    rankdir=TB;
    "raw[_tsss][_ica]" -> "high pass\nfilter";
    "raw[_tsss][_ica]" -> events;
    "raw[_tsss][_ica]" -> "[OPT: apply\nSSP vectors]" -> "raw[_tsss][_ica]";
    "high pass\nfilter" -> "extract epochs";
    events -> "extract epochs";
    "define gross\nrejection thresholds" -> "extract epochs";
    "extract epochs" -> "low pass\nfilter";
    "low pass\nfilter" ->  "clean epochs";
    "final threshold-\nbased rejection" -> "clean epochs";
    "[OPT: ICA-based\nartefact removal]" -> "clean epochs";
}
```

```python
%%dot -f svg

digraph G {
    rankdir=TB;
    "raw(_tsss)" -> "find EOG events" ->  "blink ICs" -> "define\nartefactual\ncomponents";
    "raw(_tsss)" -> "find ECG events" ->  "heart ICs" -> "define\nartefactual\ncomponents";
    ICA -> "blink ICs";
    ICA -> "heart ICs";
    "raw(_tsss)" -> ICA 
}
```
